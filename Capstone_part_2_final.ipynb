```python
from imblearn.under_sampling import NearMiss
nm = NearMiss()
X_res, y_res = nm.fit_resample(X, y)
X_res.shape, y_res.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import precision_score,recall_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=10)
```

```python
performance_metrics = pd.DataFrame(columns=['Model', 'Accuracy_Training_Set', 'Accuracy_Test_Set', 'Precision', 'Recall', 'f1_score'])
```

```python

ml_models = [LogisticRegression(max_iter=1000),  DecisionTreeClassifier(), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
models_list = [LogisticRegression(),
               DecisionTreeClassifier(),
               RandomForestClassifier(),
               #KNeighborsClassifier()
               ]
```

```python
def get_perf_metrics(model, i):
    model_name = type(model).__name__
    print("Training {} model...".format(model_name))
    model.fit(X_train, y_train)
    print("Completed {} model training.".format(model_name))
    y_pred = model.predict(X_test)


    performance_metrics.loc[i] = [
        model_name,
        model.score(X_train, y_train),
        model.score(X_test, y_test),
        precision_score(y_test, y_pred),
        recall_score(y_test, y_pred),
        f1_score(y_test, y_pred),
    ]

    print("Completed {} model's performance assessment.".format(model_name))
```

```python
from sklearn.metrics import r2_score,f1_score
for n, model in enumerate(models_list):
    get_perf_metrics(model, n)
```
